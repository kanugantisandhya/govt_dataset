```python
# Visualization 1: Distribution of Categories
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Category', order=df['Category'].value_counts().index)
plt.title('Distribution of Cosmetic Categories')
plt.show()
```

```python
# Visualization 2: Companies Count per Zone
plt.figure(figsize=(8, 5))
df['ZONE'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Number of Companies in Each Zone')
plt.show()
```

```python
 #Visualization 3: License Issue Date Trend
df['Year'] = pd.to_datetime(df['Date'], dayfirst=True).dt.year
plt.figure(figsize=(8, 5))
df['Year'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title('Number of Licenses Issued Per Year')
plt.show()
```

```python
# Visualization 4: Top 10 Companies with Most Entries
top_companies = df['Name'].value_counts().nlargest(10)
plt.figure(figsize=(10, 5))
top_companies.plot(kind='barh', color='lightcoral')
plt.title('Top 10 Companies with Most Entries')
plt.show()
```

```python
# Visualization 5: Distribution of Validity Periods
df['Validity Year'] = pd.to_datetime(df['Validity'], dayfirst=True).dt.year
df['Validity Period'] = df['Validity Year'] - df['Year']
sns.histplot(df['Validity Period'].dropna(), bins=20, kde=True)
plt.title('Distribution of License Validity Periods')
plt.show()
```

```python
# Visualization 6: Pie Chart for Category Distribution
plt.figure(figsize=(8, 8))
df['Category'].value_counts().head(5).plot.pie(autopct='%.1f%%', colors=sns.color_palette('pastel'))
plt.title('Top 5 Cosmetic Categories')
plt.ylabel('')
plt.show()
```

```python
# Visualization 8: Companies Established Before 2000
df_before_2000 = df[df['Year'] < 2000]
plt.figure(figsize=(10, 5))
sns.countplot(data=df_before_2000, y='ZONE', order=df_before_2000['ZONE'].value_counts().index)
plt.title('Companies Established Before 2000 (By Zone)')
plt.show()
```

```python
# Visualization 9: Bar Chart for Top License Numbers
plt.figure(figsize=(10, 6))
if not df['Lic No'].isna().all():
    df['Lic No'].value_counts().nlargest(10).plot(kind='bar', color='orange')
    plt.title('Top 10 License Numbers')
    plt.show()
else:
    print('No valid License Number data available for visualization.')
```

```python
# Visualization 10: Companies Without Zone Information
plt.figure(figsize=(7, 4))
df['ZONE'].isna().value_counts().plot(kind='bar', color=['green', 'red'])
plt.title('Companies with Missing Zone Information')
plt.xticks([0, 1], ['No Missing Info', 'Missing Info'])
plt.show()
```

```python
# Visualization 11: Companies with Maximum Validity Periods
df_sorted_validity = df[['Name', 'Validity Period']].nlargest(10, 'Validity Period')
plt.figure(figsize=(10, 5))
sns.barplot(data=df_sorted_validity, x='Validity Period', y='Name', palette='viridis')
plt.title('Top 10 Companies with Maximum Validity Periods')
plt.show()
```

```python

# Visualization 12: Distribution of Companies by Address
plt.figure(figsize=(12, 6))
df['Address'].value_counts().nlargest(10).plot(kind='bar', color='purple')
plt.title('Top 10 Frequent Company Addresses')
plt.show()
```

```python
# Visualization 13: Zone-wise Category Distribution
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='ZONE', hue='Category')
plt.title('Zone-wise Category Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 14: Companies Established After 2010
df_after_2010 = df[df['Year'] > 2010]
plt.figure(figsize=(10, 5))
sns.countplot(data=df_after_2010, y='ZONE')
plt.title('Companies Established After 2010 (By Zone)')
plt.show()
```

```python
# Visualization 16: Top 5 Categories with Longest Validity Period
plt.figure(figsize=(8, 5))
df.groupby('Category')['Validity Period'].mean().nlargest(5).plot(kind='bar', color='teal')
plt.title('Top 5 Categories with Longest Average Validity Period')
plt.show()
```

```python
# Visualization 17: Box Plot for Validity Period by Zone
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='ZONE', y='Validity Period')
plt.title('Box Plot for Validity Period by Zone')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualization 18: Line Chart of License Issued by Year
plt.figure(figsize=(10, 6))
df.groupby('Year').size().plot(kind='line', marker='o', color='blue')
plt.title('Trend of Licenses Issued Over the Years')
plt.show()
```

```python
# Visualization 19: Violin Plot of Validity Period
plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x='ZONE', y='Validity Period')
plt.title('Violin Plot for Validity Period by Zone')
plt.xticks(rotation=45)
plt.show()
```

```python

# Visualization 20: Companies Established by Decade
df['Decade'] = (df['Year'] // 10) * 10
plt.figure(figsize=(8, 5))
df['Decade'].value_counts().sort_index().plot(kind='bar', color='pink')
plt.title('Number of Companies Established by Decade')
plt.show()
```
